# 성별에 따른 tips 차이?
* 남성 손님들이 준 팁의 평군, 여성 손님들의 tip의 평균


```python
# select distinct sex from tips;
tips['sex'].unique()
```

```python
genders = tips.sex.unique()
```

```python
for x in genders:
  # print(x)
  tip_mean = tips[tips.sex == x].tip.mean()
  print(f'{x}: {tip_mean}')
```

```python
mean_female = tips[tips.sex == 'Female'].tip.mean()
mean_male = tips[tips.sex == 'Male'].tip.mean()

dict1 = {
    'female_tip_mean': mean_female,
    'Male_tip_mean': mean_male
}
dict1
df = pd.DataFrame(dict1, index = [1])
df
```

* 성별 tip의 최댓값

```python
for x in genders:
  tip_max = tips[tips.sex == x].tip.max()
  print(f'{x}: {tip_max}')
```

* 성별 tip의 최솟값

```python
for x in genders:
  tip_min = tips[tips.sex == x].tip.min()
  print(f'{x}: {tip_min}')
```

성별 tip의 최댓값 & 최솟값




```python
max_female = tips[tips.sex == 'Female'].tip.max()
min_female = tips[tips.sex == 'Female'].tip.min()
max_male = tips[tips.sex == 'Male'].tip.max()
min_male = tips[tips.sex == 'Male'].tip.min()

dict2 = {
    '여성 tip 최대': max_female,
    '여성 tip 최소': min_female,
    '남성 tip 최대': max_male,
    '남성 tip 최소': min_male
}
dict2
df = pd.DataFrame(dict2, index = [1])
df
```

## 2개 이상의 카테고리별 groupby
다단계 인덱스(multi label index)를 가지는 Series


```python
# 성별, 흡연여부별 tip의 평균
tips.groupby(['sex','smoker']).tip.mean()
```

```python
tips.groupby(['smoker', 'sex']).tip.mean()
```

```python
# 성별, 요일별 tip의 평균
tips.groupby(['sex', 'day']).tip.mean()
```

```python
tips.groupby(['day', 'sex']).tip.mean()
```

## 카테고리 별로 2개 이상의 변수(컬럼)의 통계값 찾기

```python
tips[['tip', 'total_bill']]
```

```python
# 성별 영수증, 팁의 평균.
tips.groupby('sex')[['tip', 'total_bill']].mean()
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균
tips.groupby(['sex', 'smoker'])[['tip', 'total_bill']].mean()
```

## 그룹별 2개 이상의 통계함수 적용

```python
# 성별 tip의 평균, 최댓값, 중앙값 ,최솟값
# agg == aggregate
tips.groupby('sex').tip.agg([np.mean, np.min, np.max, np.median])
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균과 표준편차
tips.groupby(['sex', 'smoker'])[['tip', 'total_bill']].agg([np.mean, np.std])
```

# `pivot_table` 함수(메서드)
조건에 상관없이 결과를 dataFrame으로 줌.
* Parameters:
  * index: 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
  * columns: 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
  * values: 데이터로 사용할 열/ 통계 함수를 적용할 변수(컬럼)
  * agggfunc: 데이터 집계함수(aggregation function). 기본값은 평균(mean)

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex') # aggfunc = 'mean' (기본값)
```

```python
# 성별, 흡연여부별 팁의 평균.
tips.pivot_table(values= 'tip', index= ['sex','smoker'] )
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일 별, 시간별 팁의 평균.
tips.pivot_table(values= 'tip', index=['sex', 'smoker'], columns= ['day','time'])
```

```python
# 요일별, 시간별 영수증과 팁의 평균, 표준편차
tips.pivot_table(values= ['total_bill','tip'], index=['day','time'], aggfunc= [np.mean, np.std])
```

```python
tips.pivot_table(values= ['total_bill', 'tip'], index='day', columns='time', aggfunc= [np.mean, np.std] )
```
