```python
tips[tips.total_bill == tips.total_bill.max()]
```

# 성별에 따른 tips 차이?
* 남성 손님들이 준 팁의 평군, 여성 손님들의 tip의 평균
* 성별 tip의 최댓값
* 성별 tip의 최솟값

```python
mean_female = tips[tips.sex == 'Female'].tip.mean()
mean_male = tips[tips.sex == 'Male'].tip.mean()

dict1 = {
    'female_tip_mean': mean_female,
    'Male_tip_mean': mean_male
}
dict1
df = pd.DataFrame(dict1, index = [1])
df
```

```python
max_female = tips[tips.sex == 'Female'].tip.max()
min_female = tips[tips.sex == 'Female'].tip.min()
max_male = tips[tips.sex == 'Male'].tip.max()
min_male = tips[tips.sex == 'Male'].tip.min()

dict2 = {
    '여성 tip 최대': max_female,
    '여성 tip 최소': min_female,
    '남성 tip 최대': max_male,
    '남성 tip 최소': min_male
}
dict2
df = pd.DataFrame(dict2, index = [1])
df
```

# 흡연 여부에 따른 tip의 차이?
* 흡연 여부(Yes/no) 별 tip의 평균

```python
yes = tips[tips.smoker == 'Yes'].tip.mean()
no = tips[tips.smoker == 'No'].tip.mean()

dict3 = {
    '흡연자 tip 평균': yes,
    '비흡연자 tip  평균': no
}
df = pd.DataFrame(dict3, index= [1])
df
```

# 요일에 따른 tip의 차이?

```python
tips[tips.day == 'Sat'].tip.mean()
tips[tips.day == 'Sun'].tip.mean()
tips[tips.day == 'Thur'].tip.mean()
tips[tips.day == 'Fri'].tip.mean()

dict4 = {
    'Sat tip mean': tips[tips.day == 'Sat'].tip.mean(),
    'Sun tip mean': tips[tips.day == 'Sun'].tip.mean(),
    'Thur tip mean': tips[tips.day == 'Thur'].tip.mean(),
    'Fri tip mean': tips[tips.day == 'Fri'].tip.mean(),
}
dict4
```
