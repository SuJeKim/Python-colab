## indexing, slicing
* `iloc`: integer location정수 인덱스(ndarray의 기본 인덱스) 기반으로 원소 참조 또는 슬라이싱
   * iloc에서 사용하는 index가 Series(생성자)가 자동으로 메겨진 s.index가 아님.
   * ndarray의 index.
* `loc`: location. 레이블 기반의 원소 참조 또는 슬라이싱.  

   * 끝나는 값도 포함.(포스트잇도 붙여있다고 생각하기)
   * Series(생성자)가 자동으로 메겨진 s.index사용.

```python
np.random.seed(1)
s1 = pd.Series(data=np.random.randint(-10, 10, size = 6))
s1
```

```python
# iloc를 사용한 인덱싱
print(s1.iloc[0]) # Series에서 첫번째 원소
print(s1.iloc[-1]) # Series에서 마지막 원소
```

```python
# iloc를 사용한 슬라이싱
print(s1.iloc[:3]) # Series에서 첫 3개 원소 자르기
print(s1.iloc[-3:]) # Series에서 마지막 3개원소 자르기
```

```python
# loc를 사용한 인덱싱
print(s1.loc[0]) # Series의 첫번쨰 원소
print(s1.loc[5]) # Series의 마지막 원소 - 음수 인덱스 사용 불가능!
```

```python
print(s1.loc[-1])
```

```python
# loc를 사용한 슬라이싱
print(s1.loc[0:2])
```

## DataFrame에서 조건에 맞는 행을 선택 - boolean indexing

* select * from table where col_name = ?

```python
df
```

```python
df.year == 2023
```

```python
df[df.year == 2023]
```

* boolean indexing을 사용할 때는, 파이썬의 논리 연산자(`and, or, not`)를 사용할 수 없음!
* boolean indexing에서는 `&, |, ~` 연산자를 사용함!
* 조건식과 조건식은 ()를 사용해서 구분, 연산의 순서를 명시해야 함.

```python
# city가 '서울'이거나 pop이 0.65이하인 행들을 선택:
df[(df['city'] == '서울') | (df['pop'] <= 0.65)]
# df[(df.city == '서울') | (df['pop'] <= 0.65)]
```

```python
# 경기도의 2023년 자료를 선택
df[(df.city == '경기') & (df.year == 2023)]
```

* 반별(class 별) 수학 점수 평균을 찾고 출력.

```python
# class 컬럼의 빈도수
exam['class'].value_counts()
```

```python
# 수학, 영어, 과학 컬럼의 기술 통계량을 출력.
exam[['math', 'english', 'science']].describe() # [a[b]] a: 인덱스 연산자, b: 리스트 연산자
```

```python
# 1반 학생들의 데이터를 출력.
exam[exam['class'] == 1]
```

```python
# 수학 평균을 찾고 출력.
exam['math'].mean()
```

```python
# 수학 점수 표준편차
exam['math'].std()
```

```python
# 수학 점수가 평균 이상인 학생들의 데이터를 출력
exam[exam.math >= exam.math.mean()]
```

```python
# 1반 학생들의 수학 점수 평균을 찾고 출력
exam[exam['class'] == 1]['math'].mean()
```

```python
# 2반 학생들의 수학 점수 평균을 찾고 출력.
exam[exam['class'] == 2].math.mean()
```

```python
# 반별(class 별) 수학 점수 평균을 찾고 출력.
for x in range(1, 6):
  math_mean = exam[exam['class'] == x].math.mean()
  print(f'class {x} : {math_mean}')
```

```python
means = []
for x in range(1, 6):
  math_mean = exam[exam['class'] == x].math.mean()
  means.append(math_mean)

print(means)

exam_dict = {
    'class': np.arange(1,6),
    'math_mean': means
}
print(exam_dict)

df = pd.DataFrame(exam_dict)
df
```

```python
# 세 과목의 점수가 모두 평균 이상인 학생들의 데이터를 출력.
exam[(exam.math >= exam.math.mean()) &
     (exam.english >= exam.english.mean()) &
     (exam.science >= exam.science.mean())]
```

```python
# 1반 학생들의 수학 평균 점수 이상인 학생들.
math_mean = exam[exam['class'] == 1].math.mean()
print(math_mean)
exam[exam.math >= math_mean]
```
