# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체 생성
1. `figure` 객체의 subplot(axis- 축)을 생성해서 추가
1. subplot에 그래프 그림
1. 그래프의 여러가지 옵션들을 설정
1. 그래프 보여주기

`figure`: 그림을 그릴 수 있는 스케치북

subplot: 그래프를 몇 개를 그릴건지

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가
# axis = figure.add_subplot(1,1,1) # (행의 개수, 열의 개수, 인덱스) / 인덱스의 기본값: 1
axis = figure.add_subplot(111) # 211: 행/열/인덱스
# figure.add_subplot(212)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x=tips['total_bill'])

# 4 그래프의 옵션 설정
axis.set_xlabel('Total Bill')
axis.set_ylabel('Count') # 노트북 코드 cell의 특징: 가장 마지막 코드에 리턴값/expression을  출력
axis.set_title('Histogram')
axis.grid()

# 5. 그래프 보여주기 ->  주피터 노트북 환경이 아닐경우에 보여줌
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기

```python
# 1. figure 객체 생성
fig = plt.figure(figsize=(6,8)) # (6, 8) -> 가로(6):세로(8)

# 2. subplot을 figure에 추가
axis1 = fig.add_subplot(2,1,1) # add_subplot(211)
axis2 = fig.add_subplot(2,1,2) # add_subplot(212)

# 3. 그래프
axis1.hist(x=tips.total_bill)
axis2.scatter(x= tips.total_bill, y=tips.tip)

# 4. 옵션
axis1.grid()
axis2.grid

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
# 1. figure 객체 생성
fig = plt.figure(figsize=(12,4))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(1,2,1) # add_subplot(121)
ax2 = fig.add_subplot(1,2,2) # add_subplot(222)

ax1.hist(x=tips.tip)
ax2.scatter(x=tips.total_bill, y=tips.tip)

ax1.grid()
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total bill')
ax2.set_ylabel('Tip')
ax2.set_title('Scatter plot')

plt.show()
```

# `plt.subplots()` 함수

figure 생성 + subplot 추가

# figure에 1개의 subplot이 있는 경우

Returns

fig : .Figure

ax : ~.axes.Axes or array of Axes



```python
#  figure 생성, figure에 subplot을 추가
fig, ax = plt.subplots()

# subplot에 그래프 그림
ax.hist(x=tips.total_bill)

# 옵션

# 그래프 보여주기
plt.show()
```

## subplot 2개를 상하로 배치

```python
fig, ax = plt.subplots(nrows = 2, figsize=(6,8)) #  **fig_kw: Any -> 선언되지 않는 figure()의param도 넘길 수 있음  All additional keyword arguments are passed to the.pyplot.figure call.
print(ax) # NumPy 배열 -> subplot이 2개 이상일 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

## 2개의 subplot을 좌우로 배치

```python
fig, ax = plt.subplots(ncols= 2, figsize=(12,6))
print(ax)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

## Histogram
* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있음.

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 구간
* color: 막대 안쪽을 채우는 색
* edgecolor: 막대 테두리 색
* density: False - 개수/ True - 밀도(비율)

```python
# 1. figure 생성. 2. subplot 1개 생성 3. histogram 그림.
plt.hist(x=tips.total_bill, color='purple', edgecolor='black', bins=10, density=True)

# 4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

## Bar plot(막대 그래프)

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
type(gender_count)
```

```python
gender_count.index
```

```python
gender_count.values
```

```python
# 세로 막대 그래프(vertical bar plot)
# x: 인덱스, height: 막대의 높이. 빈도수
plt.bar(x=gender_count.index, height=gender_count.values, width=0.5)

plt.show()
```

```python
# 가로 막대 그래프(horizontal bar graph)
plt.barh(y=gender_count.index, width=gender_count.values, height=0.5)

plt.show()
```

### EX 4. 성별 tip의 boxplot을 하나의 subplot에 그리기.

### EX 5. 시간별 tip의 box plot을 하나의 subplot에 그리기.

```python
times = tips.time.unique()
tip_by_time = [] # Lunch, Dinner 부분집합에서 tip 컬럼을 저장하기 위한 리스트
for x in times:
  subset = tips[tips.time == x].tip
  tip_by_time.append(subset)

# print(tip_by_time)

plt.boxplot(x=tip_by_time, labels= times)
plt.show()
```

```python
sns.boxplot(x="time", y="tip",data=tips, width= 0.5)

plt.show()
```

### 요일별 tip의 box plot을 하나의 subplot에 그리기.

```python
days = tips.day.unique()
# print(days)
tip_by_day = [ tips[tips.day == x].tip for x in days ]

plt.boxplot(x=tip_by_day, labels=days)
plt.show()
```

```python
sns.boxplot(x="day", y="tip",data=tips, width= 0.5)

plt.show()
```
